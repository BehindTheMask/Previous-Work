```python
import pandas as pd
import numpy as np
orders = pd.read_csv('olist_orders_dataset.csv')
```

```python
orders = pd.DataFrame(orders)
```

```python
orders['order_status'] = np.where(orders.order_status == 'delivered', 1, 0)
```

```python
orders = orders.drop('order_approved_at',axis = 1)
```

```python
orders = orders.drop('order_delivered_carrier_date',axis  = 1)
```

```python
orders.head(10)
```

```python
orders = orders.dropna(axis=0,how='any')
```

```python
orders
```

```python
orders = orders.drop('order_status',axis  = 1)
```

```python
orders.head
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'], dayfirst=True)
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], dayfirst=True)
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'], dayfirst=True)
```

```python
orders['order_purchase_timestamp'] = pd.DatetimeIndex(orders['order_purchase_timestamp']).date
orders['order_delivered_customer_date'] = pd.DatetimeIndex(orders['order_delivered_customer_date']).date
orders.head(10)
```

```python
orders ['delivery_time'] = orders ['order_delivered_customer_date'] - orders['order_purchase_timestamp']

```

```python
orders
```

```python
orders.dtypes
```

```python
orders['order_estimated_delivery_date'] = pd.DatetimeIndex(orders['order_estimated_delivery_date']).date
orders ['es_delivery_time'] = orders ['order_estimated_delivery_date'] - orders['order_delivered_customer_date']
```

```python
orders
```

```python
orders ['delivery_time'] = (orders.delivery_time/np.timedelta64(1,'D')).astype(int)
orders ['es_delivery_time'] = (orders.es_delivery_time/np.timedelta64(1,'D')).astype(int)
```
